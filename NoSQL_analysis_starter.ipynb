### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the query filter
query = {"BusinessName": "Penang Flavours"}

# Define the projection (only return latitude & longitude)
projection = {
    "geocode.latitude": 1,
    "geocode.longitude": 1,
    "_id": 0  # Exclude the _id field
}

# Execute the query and store the result
result = uk_food["establishments"].find_one(query, projection)

# Extract latitude and longitude as floats
if result:
    latitude = float(result.get("geocode", {}).get("latitude", 0))
    longitude = float(result.get("geocode", {}).get("longitude", 0))
else:
    latitude, longitude = None, None

print(f"For Penang Flavours, the latitude is {latitude} and the longitude is {longitude}")

```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01

query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "scores.Hygiene": {"$ne": None}  # Exclude documents where scores.hygiene is None
    
}

# Limit the fields to be returned to make result more readable

projection = {
    "BusinessName": 1,  # Include business name
    "geocode.latitude": 1,
    "geocode.longitude": 1,
    "scores.Hygiene": 1,  # Include hygiene score
    "_id": 0  # Exclude the _id field
}
   
results = list(uk_food["establishments"].find(query, projection).sort("scores.Hygiene", 1).limit(5))

#Print the results
for doc in results:
    print(doc)

```

```python
# Convert result to Pandas DataFrame
near_penang_flavours = pd.DataFrame(results)

# Display the DataFrame
near_penang_flavours
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0


# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {"$match": {"scores.Hygiene": 0}},  # Filter establishments with hygiene score = 0
    {"$group": {
        "_id": "$LocalAuthorityName",  # Group by LocalAuthorityName
        "count": {"$sum": 1}  # Count establishments
    }},
    {"$sort": {"count": -1}}  # Sort by count in descending order
]

# Execute the aggregation pipeline
results = list(uk_food["establishments"].aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(results)}")

# Print the first 10 results
print("First 10 results:")
for doc in results[:10]:  # Slice to get the first 10 documents
    print(doc)
```
